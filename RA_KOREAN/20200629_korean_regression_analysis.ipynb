```python
import pandas as pd
import numpy as np

try:
    data = pd.read_csv('all_articles_korean_vec_20060101_20061231_4.csv', encoding = 'cp949')
except UnicodeEncodeError:
    data = pd.read_csv('all_articles_korean_vec_20060101_20061231_4.csv', encoding = 'utf8')

data.head()
```

```python
stock = pd.read_csv("stock_return.csv")
stock.head()
```

```python
data.rename(columns={'ART_ID': 'news_id'}, inplace=True)
data.head()
```

```python
print(data.shape, stock.shape)
```

```python
data_joined = pd.merge(data, stock, left_on='news_id', right_on='news_id', how='inner')
data_joined.head()
```

```python
data_joined.shape
```

```python
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

model = linear_model.LinearRegression()
```

```python
x_vars = data_joined.columns.tolist()[1:-3]
print(x_vars)
```

```python
model.fit(data_joined[x_vars], data_joined.ret)
print(model.coef_, model.intercept_)
```

```python
temp = ""
for i in range(0, 1000):
    temp += x_vars[i]
    temp += "+"
temp = temp[:-1]
```

```python
import statsmodels.formula.api as sm
#from statsmodels.sandbox.regression.predstd import wls\_prediction\_std

result = sm.ols(formula = 'ret ~ '+ temp, data = data_joined).fit()
result.summary()
```
