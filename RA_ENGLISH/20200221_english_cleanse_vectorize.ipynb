```python
import pandas as pd
import numpy as np
import re
import os

import nltk
import re
import gensim
import itertools

import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from sklearn.feature_extraction.text import CountVectorizer

nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_word = stopwords.words('english')
stop_word.extend(['from', 're', 'use'])
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc = True))
```

```python
extracted_path = []
extracted_name = []

def file_folder_split_csv(dir):
    for x in os.listdir(dir):
        if os.path.isfile(os.path.join(dir, x)) and "csv" in x:
            extracted_path.append(os.path.join(dir, x))
            extracted_name.append(x)
        if os.path.isdir(os.path.join(dir, x)):
            new_dir = os.path.join(dir, x)
            file_folder_split_csv(os.path.join(dir, x))

file_folder_split_csv('./articles_english_year_month') 
print(extracted_path[0])
```

```python
#Cleansing
for data in extracted_path:
    print(data)
    articles = pd.read_csv(data)
    #articles = articles[['ART_ID', 'ART_CONTENT', 'MONTH']].dropna(axis = 0)
    #articles = articles.reset_index(drop = True)

    text = articles['article'].values.tolist()
    pre_text = [re.sub('\S*@\S*\s?',' ', sent) for sent in text]
    pre_text = [re.sub('\s+', ' ', sent) for sent in text]
    pre_text = [re.sub('\'', ' ', sent) for sent in text]
    
    filtered = []
    for sentence in pre_text:
        tokens = word_tokenize(sentence)
    
        result = []
        for w in tokens:
            if w not in stop_word:
                result.append(w)
    
        filtered.append(result)
    
    data_words = list(sent_to_words(filtered))
    articles['article_cleaned'] = data_words

    articles.to_csv(data, index = False)
```

```python
temp = []
temp2 = []
def file_folder_split_temp(dir):
    data = []
    for x in os.listdir(dir):
        if os.path.isfile(os.path.join(dir, x)) and ("csv" in x):
            data.append(os.path.join(dir, x))
            temp2.append(x)
        if os.path.isdir(os.path.join(dir, x)):
            file_folder_split_temp(os.path.join(dir, x))
    
    if data:
        temp.append(data)
file_folder_split_temp('./articles_english_year_month')
```

```python
data_index = 0
for data_list in temp:
    partial_articles = []
    year = data_list[0].split('\\')[1]
    month = data_list[0].split('\\')[2]
    print(year)
    for data in data_list:
        articles = pd.read_csv(data)
        articles['article_cleaned'] = articles['article_cleaned'].str.replace('[', '')
        articles['article_cleaned'] = articles['article_cleaned'].str.replace(']', '')
        articles['article_cleaned'] = articles['article_cleaned'].str.replace('\'', '')
        partial_articles.append(articles)
        data_index += 1
    
    all_articles = partial_articles[0]
    for i in range(1, len(partial_articles)):
        all_articles = pd.concat([all_articles, partial_articles[i]])
    
    all_articles.to_csv('all_articles_english_' + str(year) + '_' + str(month) + '.csv', index = False)
    all_id = all_articles[['company', 'gvkey', 'accessNo']]

    vec = CountVectorizer()
    text_vec = vec.fit_transform(all_articles.article_cleaned.values.astype(str))
    count_vec_df = pd.DataFrame(text_vec.todense(), columns = vec.get_feature_names())
    all_id = all_id.reset_index(drop=True)
    yearly_vec = pd.concat([all_id, count_vec_df], axis = 1)
    yearly_vec.to_csv('all_articles_english_vec_' + str(year) + "_" + str(month) + '.csv', index = False)
```
